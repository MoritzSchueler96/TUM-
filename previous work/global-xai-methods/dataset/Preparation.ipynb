# Dataset preparation

We agreed on removing duplicates from the dataset and remove rows we don't need.

```python
# Load the dataset
#dataset_path = '/home/simon/Downloads/crowdtangle/2021-02-16-21-36-49-CET-Historical-Report-RT-2018-05-31--2020-12-31.csv'
dataset_path = '/home/simon/Downloads/2021-02-03-11-39-15-CET-Historical-Report-conspiracy-list-hand-picked-2020-06-30--2020-10-01.csv'
output = '/home/simon/Downloads/xxx.csv'
```

```python
df = pd.read_csv(dataset_path, low_memory=False)
df_len = len(df)
df
```

```python
"before:", df_len, "after:", len(df.drop_duplicates(subset=['Message', 'Image Text']))
```

```python
df = df.assign(
        reactions_sum=df.apply(lambda r: r['Likes'] + r['Comments'] + r['Shares'] + r['Love'] + r['Wow'] + r['Haha'] + r['Sad'] + r['Angry'] + r['Care'], axis=1).values
)
df = df.assign(
        reactions_groupavg=df.groupby('Facebook Id')['reactions_sum'].transform('mean').values
)
df = df.assign(
        score=df.apply(lambda r: (r['reactions_sum'] / r['reactions_groupavg']), axis=1).values
)
df
```

```python
df.to_csv(output)
```
